## Merge and Clean Trip History Data

#### Import Dependencies

```python
import pandas as pd
import os
from glob import glob
```

#### Read Data File 

```python
#Assigned a variable for all the csv files 
DataBikeTripFiles = ["RawData/CityBikeTripData/JC-201801-citibike-tripdata.csv.zip", "RawData/CityBikeTripData/JC-201802-citibike-tripdata.csv.zip","RawData/CityBikeTripData/JC-201803-citibike-tripdata.csv.zip","RawData/CityBikeTripData/JC-201804-citibike-tripdata.csv.zip","RawData/CityBikeTripData/JC-201805-citibike-tripdata.csv.zip","RawData/CityBikeTripData/JC-201806-citibike-tripdata.csv.zip","RawData/CityBikeTripData/JC-201807-citibike-tripdata.csv.zip","RawData/CityBikeTripData/JC-201808-citibike-tripdata.csv.zip","RawData/CityBikeTripData/JC-201809-citibike-tripdata.csv.zip","RawData/CityBikeTripData/JC-201810-citibike-tripdata.csv.zip","RawData/CityBikeTripData/JC-201811-citibike-tripdata.csv.zip","RawData/CityBikeTripData/JC-201812-citibike-tripdata.csv.zip"]
```

#### Iterate through files to create a single dataframe with data for each month

```python
dataframe = []
for d in DataBikeTripFiles:
    dataframe.append(pd.read_csv(d))
BikeTripDataframe = pd.concat(dataframe, ignore_index=True)
```

```python
BikeTripDataframe.head()
```

#### Replace 1 and 2 for Male and Female Value in the Gender Column

```python
BikeTripDataframe["gender"].replace([1, 2], ["Male","Female"], inplace=True)
```

```python
BikeTripDataframe.head()
```

#### Add Current Year Column to DataFrame

```python
BikeTripDataframe.insert(14, "current year", 2018, True) 
```

```python
BikeTripDataframe
```

#### Add Age Column to DataFrame

```python
BikeTripDataframe['age'] = BikeTripDataframe['current year'] - BikeTripDataframe['birth year'] 
```

#### Remove Values with birth year 1887 and no gender 

```python
BikeTripDataframe.loc[BikeTripDataframe['birth year'] == 1887]
```

#### Update BikeTripDataframe to not include values where birth year is 1887

```python
BikeTripDataframe = BikeTripDataframe[BikeTripDataframe['birth year']!= 1887]
```

```python
BikeTripDataframe = BikeTripDataframe[BikeTripDataframe['birth year']!= 1888]
```
